### 1.1 기본 영상 표시와 종료

'bird.png' 파일을 읽고 윈도우에 표시하며, 'q'로 종료하는 프로그램을 작성하라.
단, 지난 실습 시간에 만들었던 load_image 함수를 사용하고, main body에서는 예외 처리를 해야 한다.

````python
import os
import cv2

def load_image(filename): 
    if not os.path.exists(filename): #지정한 이미지 파일이 디스크에 존재하지 않을 경우
        raise FileNotFoundError(f"File {filename} does not exist.")   #raise는 글로벌변수처럼 사용
    img = cv2.imread(filename) #지역변수
    if img is None: #이미지 파일이 존재하더라도 cv2.imread()가 정상적으로 이미지를 읽어오지 못한 경우
        raise ValueError(f"Could not load {filename}")
    return img

try: 
    img = load_image('bird.png') #글로벌변수
    cv2.namedWindow('Image', cv2.WINDOW_AUTOSIZE) #Image라는 이름의 창, 이미지 크기에 맞게 창 크기를 자동 조절
    while True:
        cv2.imshow('Image', img)
        if cv2.waitKey(1) & 0xFF == ord('q'): #1초동안 키 입력 기다림 & 일부 운영체제에서 키 값이 이상하게 나오는걸 막기 == q 누르면 break
            break
        
except FileNotFoundError as e:
    print(f"Error : {str(e)}")
    
except ValueError as e:
    print(f"Error : {str(e)}")
    
finally:
    cv2.destroyAllWindows()    
````

### 1.2 type hint 사용하기

1. 타입 힌트 개요
- Python의 타입 힌트(Type Hint)는 코드에서 변수, 함수 매개변수, 반환값 등의 데이터 타입을 명시할 수 있게 해주는 기능
- Python 3.5부터 도입
- 주로 코드의 가독성을 높이고, 개발 도구(예: IDE, 정적 분석 도구)를 활용해 타입 오류를 사전에 잡아내는 데 사용됨

2. 타입 힌트란?
- Python은 기본적으로 동적 타입 언어로, 변수 선언 시 타입을 명시하지 않아도 됨
- 그러나 타입 힌트를 사용하면 코드에 "이 변수나 함수는 어떤 타입을 기대한다"는 정보를 추가할 수 있음
- 목적
  * 코드 문서화: 타입을 명시해 읽기 쉽게 만듦.
  * 오류 방지: 개발 중 타입 불일치를 미리 감지.
  * 도구 지원: PyCharm, VSCode 같은 IDE에서 자동 완성과 타입 체크 제공.
- 특징
  * 타입 힌트는 실행 시 강제되지 않음 (즉, 런타임에 영향을 주지 않음).
  * 단지 힌트일 뿐이며, 정적 타입 검사 도구(예: mypy)로만 강제할 수 있음

3. 기본 사용법
- 타입 힌트는 변수 선언, 함수 매개변수, 반환값에 적용할 수 있음.

4. 기본 문법

(1) 변수에 타입 힌트

```python
x: int = 10
name: str = "Alice"
numbers: list = [1, 2, 3]
```
 - 뒤에 타입을 명시.
 - 초기화 값은 선택 사항(없어도 됨: x: int).
(2) 함수에 타입 힌트

```python
def add(a: int, b: int) -> int:
    return a + b
```
 - 매개변수: a: int, b: int로 타입 지정.
 - 반환값: -> int로 반환 타입 지정.

(3) 복잡한 타입
typing 모듈을 사용해 더 구체적인 타입을 표현할 수 있음

```python
from typing import List, Dict, Optional

def greet(names: List[str]) -> str:
    return ", ".join(names)

def get_score(student: Dict[str, int]) -> Optional[int]:
    return student.get("score")  # None 가능
```

 - List[str]: 문자열 리스트.
 - Dict[str, int]: 키는 문자열, 값은 정수인 딕셔너리.
 - Optional[int]: int 또는 None을 반환.

````python
import os
import cv2
import numpy as np

#타입 힌트
def load_image(filename: str) -> np.ndarray: #filename:str 문자열을 받음 / Numpy배열을 반환
    if not os.path.exists(filename):
        raise FileNotFoundError(f"File {filename} does not exist.")
    img: np.ndarray = cv2.imread(filename) #img:np.ndarray 넘파이 배열이라는걸 명시
    if img is None:
        raise ValueError(f"Could not load {filename}")
    return img

try: 
    img = load_image('bird.png') #글로벌변수
    cv2.namedWindow('Image', cv2.WINDOW_AUTOSIZE) #Image라는 이름의 창, 이미지 크기에 맞게 창 크기를 자동 조절
    while True:
        cv2.imshow('Image', img)
        if cv2.waitKey(1) & 0xFF == ord('q'): #1초동안 키 입력 기다림 & 일부 운영체제에서 키 값이 이상하게 나오는걸 막기 == q 누르면 break
            break
        
except FileNotFoundError as e:
    print(f"Error : {str(e)}")
    
except ValueError as e:
    print(f"Error : {str(e)}")
    
finally:
    cv2.destroyAllWindows()    
````

````python
def load_image(filename : str) -> np.ndarray:
    """ 
    영상 파일을 읽고 오류를 체크하는 함수.
    Args:
        filename (str) : 읽을 영상 파일의 경로와 이름.
    
    Returns:
        numpy.ndarray : 읽은 영상 데이터 (BRG 형식).
    
    Raises:
        FileNotFoundError : 파일이 존재하지 않을 경우
        ValueError :파일을 읽을 수 없는 경우 (예:손상된 파일)
    """
    if not os.path.exists(filename):
        raise FileNotFoundError(f"File {filename} does not exist.")
    img: np.ndarray = cv2.imread(filename) 
    if img is None:
        raise ValueError(f"Could not load {filename}")
    return img
````

### 2.1 [단계 1] 마우스 이벤트 연결과 클릭 감지

문제 1에서 작성한 프로그램을 확장하여, 마우스 이벤트를 처리하는 콜백 함수를 추가하라. 이 함수는 왼쪽 마우스 버튼이 눌릴 때(cv2.EVENT_LBUTTONDOWN) 클릭된 위치의 좌표(x, y)를 콘솔에 출력해야 한다. 콜백 함수의 이름은 mouse_callback로 하고, 함수 시그니처는 아래를 따라야 한다.

```python
mouse_callback(event, x, y, flags, param)
```

프로그램이 실행되면 창에 영상이 표시되고, 마우스 클릭 시 콘솔에 좌표가 출력되도록 한다.

````python
for e in dir(cv2):
    if 'MOUSE' in e:
        print(e)
````

````python
import cv2
import image_tools

def mouse_callback(event, x, y, flags, param):
    """
    마우스 이벤트를 처리하는 콜백 함수
    
     Args:
        event: 마우스 이벤트 종류 (예: 클릭, 이동 등)
        x, y: 마우스 좌표 (픽셀 단위)
        flags: 키보드/마우스 상태 (Ctrl, Shift 등)
        param: setMouseCallback에서 넘긴 추가 데이터
    """
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"Clicked at ({x}, {y})")
    
IMAGE_FILE = 'bird.png' 
WINDOW_MAIN = 'Image'

try: 
    img = image_tools.load_image(IMAGE_FILE) 
    cv2.namedWindow(WINDOW_MAIN, cv2.WINDOW_AUTOSIZE)
    cv2.setMouseCallback(WINDOW_MAIN, mouse_callback)
    while True:
        cv2.imshow('Image', img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
except FileNotFoundError as e:
    print(f"Error : {str(e)}")
    
except ValueError as e:
    print(f"Error : {str(e)}")
    
finally:
    cv2.destroyAllWindows()   
````

### 2.4 [단계 2] 드래그 시작과 상태 관리

문제 2.1에서 만든 프로그램을 다음과 같이 수정하라.
- drawing 변수를 만들고 False로 초기화한다. 만약에 기존에 화면에 사각형이 있다면 사각형을 지운다.
- 왼쪽 마우스 클릭 시 그 위치를 저장한다.
- 왼쪽 마우스를 클릭한 상태에서 마우스를 drawing을 True로 설정하고 사각형을 지속적으로 업데이트한다
- 클릭한 왼쪽 마우스를 놓으면 drawing을 False로 하고 사각형 업데이트를 중단하고 최종 사각형을 표시한다.
- mouse_callback 함수의 이름을 draw_rectangle로 수정하라.

````python
import cv2
import numpy as np
import image_tools

def mouse_callback(event, x, y, flags, param):
    """마우스 이벤트를 처리하는 콜백 함수"""
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"Clicked at ({x}, {y})") #클릭한 좌표
        
def draw_rectangle(event, x, y, flags, param):
    global start_x, start_y, drawing, img_copy #글로벌변수 쓰면 모듈화가 안됨, 안쓰는게 좋음
    
    if event == cv2.EVENT_LBUTTONDOWN: #왼쪽버튼 누르면
        drawing = True
        start_x, start_y = x, y #시작 좌표 저장
        img_copy = img.copy() #원본 이미지를 복사해서, 사각형을 복사본 위에 그림
    
    elif event == cv2.EVENT_MOUSEMOVE: #마우스 움질일 때
        if drawing:
            img_copy = img.copy()
            cv2.rectangle(img_copy, (start_x, start_y), (x,y), (0,255,0), 2) #start_x에서 x까지, start_y에서 y까지
            cv2.imshow("Image", img_copy) #이미지를 화면에 표시
            
    elif event == cv2.EVENT_LBUTTONUP: #오른쪽 버튼 누르면
        drawing = False #사각형
        x1,x2 = min(start_x, x), max(start_x,x) #x1 최소, x2 최대
        y1,y2 = min(start_y,y), max(start_y, y) #y1 최소, y2 최대
        x1,y1 = max(0,x1), max(0,y1) #x1과 y1이 음수일 경우 0으로 보정
        x2,y2 = min(img.shape[1], x2), min(img.shape[0], y2) #img.shape[1] 이미지의 가로 너비, image.shape[0] : 이미지의 세로 너비
    
IMAGE_FILE = 'bird.png' 
WINDOW_MAIN = 'Image'

try: 
    img = image_tools.load_image(IMAGE_FILE) 
    cv2.namedWindow(WINDOW_MAIN, cv2.WINDOW_AUTOSIZE)
    
    img_copy = img.copy()
    drawing = False 
    start_x = -1
    start_y = -1
    
    cv2.setMouseCallback(WINDOW_MAIN, draw_rectangle)
    while True:
        cv2.imshow('Image', img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
except FileNotFoundError as e:
    print(f"Error : {str(e)}")
    
except ValueError as e:
    print(f"Error : {str(e)}")
    
finally:
    cv2.destroyAllWindows()   
````

마우스가 움직이면 시작위치와 현재 마우스 위치 사이에 초록색 사각형을 그림
마우스 떼면 좌표 순서를 정리함
잘라낸 이미지를 보여줌

### 2.6 콜백 함수 params에 딕셔너리를 전달하는 방법

````python
import cv2
import image_tools

#위에 코드에서 딕셔너리로 바꾸기
def draw_rectangle(event, x, y, flags, param):
    state = param  # param 딕셔너리 받아오기

    if event == cv2.EVENT_LBUTTONDOWN:
        state['drawing'] = True
        state['start_x'], state['start_y'] = x, y
        state['img_copy'] = state['img'].copy()

    elif event == cv2.EVENT_MOUSEMOVE:
        if state['drawing']:
            state['img_copy'] = state['img'].copy()
            cv2.rectangle(state['img_copy'], 
                          (state['start_x'], state['start_y']), 
                          (x, y), 
                          (0, 255, 0), 2)
            cv2.imshow("Image", state['img_copy'])

    elif event == cv2.EVENT_LBUTTONUP:
        state['drawing'] = False

        x1, x2 = min(state['start_x'], x), max(state['start_x'], x)
        y1, y2 = min(state['start_y'], y), max(state['start_y'], y)
        x1, y1 = max(0, x1), max(0, y1)
        x2, y2 = min(state['img'].shape[1], x2), min(state['img'].shape[0], y2)

        cropped = state['img'][y1:y2, x1:x2]
        cv2.imshow("Cropped", cropped)

# 이미지 불러오기
img = image_tools.load_image('bird.png')

# 상태 딕셔너리 만들기
state = {
    'img': img,
    'img_copy': img.copy(),
    'drawing': False,
    'start_x': -1,
    'start_y': -1
}

cv2.namedWindow('Image', cv2.WINDOW_AUTOSIZE)
cv2.setMouseCallback('Image', draw_rectangle, state)  # state를 param으로 전달

while True:
    cv2.imshow('Image', state['img'])
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
````

### 2.7 클래스를 이용하는 방법

````python
import cv2
import numpy as np
import image_tools

class ImageDrawer:
    def __init__(self, img): #생성자
        self.img = img
        self.img_copy = img.copy()
        self.drawing = False
        self.start_x = -1
        self.start_y = -1

    def draw_rectangle(self, event, x, y, flags, param): #마우스 콜백 함수
        if event == cv2.EVENT_LBUTTONDOWN: #마우스 왼쪽 누르면
            self.drawing = True
            self.start_x, self.start_y = x, y
            self.img_copy = self.img.copy()

        elif event == cv2.EVENT_MOUSEMOVE: #마우스 움직이면
            if self.drawing:
                self.img_copy = self.img.copy()
                cv2.rectangle(self.img_copy, (self.start_x, self.start_y), (x, y), (0, 255, 0), 2)
                cv2.imshow("Image", self.img_copy)

        elif event == cv2.EVENT_LBUTTONUP: #마우스 떼면
            self.drawing = False
            x1, x2 = min(self.start_x, x), max(self.start_x, x)
            y1, y2 = min(self.start_y, y), max(self.start_y, y)
            x1, y1 = max(0, x1), max(0, y1)
            x2, y2 = min(self.img.shape[1], x2), min(self.img.shape[0], y2)
            
            cropped = self.img[y1:y2, x1:x2]
            cv2.imshow("Cropped", cropped)


IMAGE_FILE = 'bird.png'
WINDOW_MAIN = 'Image'

try:
    img = image_tools.load_image(IMAGE_FILE)
    drawer = ImageDrawer(img) #클래스 인스턴스 생성

    cv2.namedWindow(WINDOW_MAIN, cv2.WINDOW_AUTOSIZE) #OpenCV창 만들기
    cv2.setMouseCallback(WINDOW_MAIN, drawer.draw_rectangle) #마우스 콜백 등록

    while True:
        cv2.imshow(WINDOW_MAIN, drawer.img) #이미지 표시
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

except FileNotFoundError as e:
    print(f"Error: {str(e)}")
except ValueError as e:
    print(f"Error: {str(e)}")
finally:
    cv2.destroyAllWindows()

        
````
