## 문제 1. 다음 프로그램애 여러 도형을 추가하라.

도형은 'd', 'e', 's', 'r' 등 다양하게 정의한 것을 여기에 표시하라.


```python
#r : 사각형(rectangle)
#s : 별(star)
#d : 다이아몬드(diamond)
#e : 타원(ellipse)
```

```python
import cv2
import numpy as np
import math
import image_tools

class ImageDrawer:
    def __init__(self, img):
        self.original_img = img
        self.img = img.copy()
        self.drawing = False
        self.start_point = (-1, -1)
        self.end_point = (-1, -1)
        self.shape = 'rectangle'  # 기본 도형은 사각형
        self.shapes = []

    def draw_rectangle(self, event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            self.drawing = True
            self.start_point = (x, y)

        elif event == cv2.EVENT_MOUSEMOVE and self.drawing:
            self.end_point = (x, y)
            self.update_shape()

        elif event == cv2.EVENT_LBUTTONUP:
            self.drawing = False
            self.end_point = (x, y)
            self.update_shape()
            self.save_shape()

    def get_star_points(self, center_x, center_y, width, height):
        points = []
        angle = math.pi / 5
        for i in range(10):
            r = width // 2 if i % 2 == 0 else width // 4
            x = int(center_x + r * math.cos(i * angle - math.pi / 2))
            y = int(center_y + r * math.sin(i * angle - math.pi / 2))
            points.append((x, y))
        return np.array(points, np.int32)


    def apply_bitwise_not(self):
        self.img = cv2.bitwise_not(self.img)


    def update_shape(self):
        self.img = self.original_img.copy()
        if self.shape == 'rectangle':
            cv2.rectangle(self.img, self.start_point, self.end_point, (0, 255, 0), 2)
            
        elif self.shape == 'star':
            cx = (self.start_point[0] + self.end_point[0]) // 2
            cy = (self.start_point[1] + self.end_point[1]) // 2
            w = abs(self.end_point[0] - self.start_point[0])
            h = abs(self.end_point[1] - self.start_point[1])
            points = self.get_star_points(cx, cy, w, h)
            cv2.polylines(self.img, [points], isClosed=True, color=(0, 0, 255), thickness=2)
            
        elif self.shape == 'diamond':
            cx = (self.start_point[0] + self.end_point[0]) // 2
            cy = (self.start_point[1] + self.end_point[1]) // 2
            w = abs(self.end_point[0] - self.start_point[0]) // 2
            h = abs(self.end_point[1] - self.start_point[1]) // 2

            points = np.array([
                (cx, cy - h),     # top
                (cx + w, cy),     # right
                (cx, cy + h),     # bottom
                (cx - w, cy)      # left
            ], np.int32)

            cv2.polylines(self.img, [points], isClosed=True, color=(0, 255, 255), thickness=2)

        elif self.shape == 'ellipse':
            cx = (self.start_point[0] + self.end_point[0]) // 2
            cy = (self.start_point[1] + self.end_point[1]) // 2
            axes = (abs(self.end_point[0] - self.start_point[0]) // 2,
                    abs(self.end_point[1] - self.start_point[1]) // 2)
            cv2.ellipse(self.img, (cx, cy), axes, 0, 0, 360, (255, 255, 0), 2)
            
    def save_shape(self):
        self.shapes.append({
            'type': self.shape,
            'start': self.start_point,
            'end': self.end_point
        })

IMAGE_FILE = 'bird.jpg'
WINDOW_MAIN = 'Image'

try:
    img = image_tools.load_image(IMAGE_FILE, 'BGR')
    drawer = ImageDrawer(img)

    cv2.namedWindow(WINDOW_MAIN)
    cv2.setMouseCallback(WINDOW_MAIN, drawer.draw_rectangle)

    while True:
        cv2.imshow(WINDOW_MAIN, drawer.img)
        key = cv2.waitKey(1) & 0xFF

        if key == ord('q'):
            break
        elif key == ord('r'):
            drawer.shape = 'rectangle'
        elif key == ord('s'):
            drawer.shape = 'star'
        elif key == ord('d'):
            drawer.shape = 'diamond'
        elif key == ord('e'):
            drawer.shape = 'ellipse'
        elif key == ord('n'):
            drawer.apply_bitwise_not()

    cv2.destroyAllWindows()

except Exception as e:
    print(f'에러 발생: {e}')
```
