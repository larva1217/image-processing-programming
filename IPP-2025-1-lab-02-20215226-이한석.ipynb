https://www.adobe.com/kr/creativecloud/file-types/image/comparison/jpeg-vs-png.html

**답**

```python
JPEG 파일 
-디지털 사진을 저장하는 가장 일반적인 형식
-압축을 통해 이미지 파일 크기를 줄여서 저장
-웹페이지에 쉽게 업로드 가능
-최대 1600만개의 색상을 담아낼수 있음
-손실압축방식, 파일이 압축될 때 이미지의 일부 데이터가 영구적으로 삭제됨
-투명한 배경 지원 안함

PNG 파일
-JPEG파일보다 저장공간을 더 많이 차지
-고품질 사진보다는 웹그래픽, 로고, 차트, 일러스트레이션에 사용
-투명한 배경이 있는 그래픽을 처리할 수 있다는 장점
-무손실 압축 방식, 이미지가 압축될 떄 손실되는 데이터가 없음
```

### 문제 5.

두 영상 $f_1(x,y)$와 $f_2(x,y)$에 각각 가중치 $\alpha$와 $1-\alpha$를 곱한 후에 더한 영상 $g(x,y)$는 다음과 같이 표현할 수 있다.

$$g(x,y) = \alpha f_1(x,y) + (1-\alpha )f_2(x,y), \;\; 0 \le \alpha \le 1$$

tree-1.jpg와 tree-3.jpg에 각각 0.5를 곱해서 더한 파일 파일을 아래 콘솔에 표시하라.

연산 결과 영상을 다시 uint8로 변환하여 화면에 표시하여야 한다.

```python
tree = 0.5*tree1+0.5*tree3 #tree1과 tree3의 픽셀 값을 가각 50%씩 더해서 새로운 이미지 만들기
#float64로 변환됨
tree = tree.astype('uint8') #정수형으로 변환
```

```python
plt.imshow(tree)
plt.show()
```

## 문제 6.
OpenCV의 imread() 함수를 이용해서 tree-1.jpg와 tree-2.jpg 두 파일을 읽고, 
가중치 $\alpha$를 1부터 0까지 0.01 간격으로 변경해 가면서, 두 영상의 weighted sum을 구하고 화면에 표시하는 프로그램을 작성하라.
단, 영상과 영상 사이의 디스플레이 간격은 0.05초로 설정하라.


```python
%matplotlib qt 
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

tree1 = cv2.imread('tree-1.jpg') #이미지 읽기
tree2 = cv2.imread('tree-2.jpg')
tree3 = cv2.imread('tree-3.jpg')
tree4 = cv2.imread('tree-4.jpg')

# figure 생성
fig, ax = plt.subplots() #fig:전체 그래프 영역, ax:이미지가 표시될 영역

img_display = ax.imshow(tree1.astype(np.uint8)) #처음 이미지

# 슬라이더 추가
ax_slider = plt.axes([0.3, 0.05, 0.4, 0.03], facecolor='lightgray')
slider = Slider(ax_slider, 'Position', 0, 399, valinit=0, valstep=1) #0부터 399까지


def update(val):
    pos = int(slider.val)  # 현재 슬라이더 값 
    
    # 각 이미지 간의 변화 단계 (0~133, 134~266, 267~399)
    if pos <= 133:
        alpha = pos / 133  # tree1 → tree2
        blended = (1 - alpha) * tree1 + alpha * tree2 #tree1에서 tree2로 부드럽게 변환
    elif pos <= 266:
        alpha = (pos - 133) / 133  # tree2 → tree3
        blended = (1 - alpha) * tree2 + alpha * tree3
    else:
        alpha = (pos - 266) / 133  # tree3 → tree4
        blended = (1 - alpha) * tree3 + alpha * tree4

    
    img_display.set_data(np.clip(blended, 0, 255).astype(np.uint8)) #blended를 0~255범위로 제한
    fig.canvas.draw_idle() #matplotlib이 화면을 즉시 업데이트하도록 설정

# 슬라이더 변경 시 update 함수 호출
slider.on_changed(update)

plt.show() #그래프 표시
```
