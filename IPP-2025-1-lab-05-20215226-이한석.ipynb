### 문제 3.

'bird.png' 파일을 읽고 중앙에 영상의 정 중앙을 원의 중심으로 하고 영상의 높이의 1/3을 반지름으로 하는 원 외부의 영역을 모두 Cyan색으로 만드는 프로그램을 작성하라. 단, OpenCV의 fillPoly 함수를 사용하여 해결하라.

```python
import numpy as np
import cv2
from image_tools import load_image

img = load_image("bird.png", color_channel="BGR")  
bird = (img * 255).astype(np.uint8)
```

```python
height, width = bird.shape[:2] #높이, 너비

CV2_COLOR_CYAN = (255, 255, 0) #색깔
center = (height // 2, width // 2)  # 중심좌표
radius = height // 3  # 반지름
center,radius
```

```python
num_points = 100 #원을 그릴 꼭짓점의 개수

angles = np.linspace(0, 2 * np.pi, num_points) #0부터 2파이까지 100간격 배열

circle_points = np.array([
    [center[0] + radius * np.cos(theta), center[1] + radius * np.sin(theta)]
    for theta in angles #각 각도마다 원의 점 계산
], dtype=np.int32)
```

```python
mask = np.zeros((height, width), dtype=np.uint8)
#height,width를 갖는 0으로 채워진 배열

cv2.fillPoly(mask, [circle_points], 255)  # 원 내부를 255로 채움
#mask배열에서 원 내부는 255, 나머지는 0

# 원 외부를 Cyan으로 채우기 위해 전체 이미지를 Cyan으로 설정
cyan_img = np.full_like(img, (255, 255, 0))  # Cyan: BGR (0, 255, 255)


result = np.where(mask[:, :, np.newaxis] == 255, img, cyan_img)
#mask[:, :, np.newaxis]==255 원 내부 찾는 조건
#원 내부는 원본 이미지
#원 외부는 시안색상

# 결과 표시 (선택 사항)
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 문제 4.
문제 2와 3을 각각 키보드 입력을 통해서 원의 크기를 확대하거나 축소하는 프로그램을 작성하라.

```python
#문제 2번
import numpy as np
import cv2
from image_tools import load_image

img = load_image("bird.png", color_channel="BGR") #img는 0~1 사이의 실수 값으로 이루어진 넘파이배열
bird = (img*255).astype(np.uint8) #img*255 실수로 표현된 값을 0~255로 바꿈, OpenCV는 uint8타입을 사용
height,width=bird.shape[:2]

CV2_COLOR_CYAN = (255,255,0)
center=(height//2,width//2) #중심좌표
radius=height//3 #반지름
center, radius

while True:
    # 좌표 배열 만들기
    x = np.arange(height).reshape(-1, 1)  # shape: (height, 1)
    y = np.arange(width).reshape(1, -1)   # shape: (1, width)

    r2 = radius ** 2
    mask = (x - center[0]) ** 2 + (y - center[1]) ** 2 > r2  # shape: (height, width)

    img_result = bird.copy()  # 이미지 복사
    img_result[mask] = CV2_COLOR_CYAN  # True인 위치만 Cyan 색상으로 변경

    # 결과 이미지 띄우기
    cv2.imshow("window", img_result)

    # 키보드 입력 처리
    key = cv2.waitKey(0) & 0xFF  # 키 입력 받기

    if key == ord('+'):  # '+' 키를 눌러 원 크기 확대
        radius += 10
    elif key == ord('-'):  # '-' 키를 눌러 원 크기 축소
        radius = max(10, radius - 10)  # 반지름은 최소 10으로 유지
    elif key == ord('q'):  # 'q' 키를 눌러 프로그램 종료
        break

cv2.destroyAllWindows()  # 창 닫기
```

## OpenCV의 resize 함수가 제공하는 보간 방법
- INTER_NEAREST: 최근접(nearest-neighbor) 보간. 처리 시간이 가장 짧음.
- INTER_LINEAR: 양방향 선형(bilinear) 보간, 기본값 
- INTER_AREA: 화소 영역의 관계를 이용한 리샘플링. 영상 축소에 주로 사용되고, 영상 확대에 사용하면 INTER_NEAREST 방법과 유사한 효과
- INTER_CUBIC: 4x4 영역에서 양방향 cubic spline 보간.
- INTER_LANCZOS4: 8x8 영역에서 Lanczos 보간. 처리 시간이 가장 느림.

### 문제 5

'clock_130.jpg' 영상 파일을 읽고, OpenCV의 윈도우를 만들어 영상을 표시하고, 마우스 입력 또는 키보드 입력을 사용하여 영상을 확대하거나 축소하는 프로그램을 작성하라.

```python
import cv2
import numpy as np

img = cv2.imread('clock_130.jpg') #이미지 읽어오기

# 초기 설정
scale = 1.0
scale_step = 0.1  # 확대/축소 비율
min_scale = 0.1 #최소
max_scale = 5.0 #최대

window_name = "Image"
cv2.namedWindow(window_name)

# 확대/축소 함수
def resize_image(img, scale):
    h, w = img.shape[:2] #이미지의 높이, 너비
    new_size = (int(w * scale), int(h * scale)) #새로운 크기 계산
    return cv2.resize(img, new_size, interpolation=cv2.INTER_LINEAR) #interpolation=cv2.INTER_LINEAR 확대/축소 시 픽셀을 부드럽게

while True:
    scaled_img = resize_image(img, scale) #이미지 스케일에 맞게 조절
    cv2.imshow(window_name, scaled_img)

    key = cv2.waitKey(0)

    if key == 27:  # ESC → 종료
        break
    elif key == ord('+'):  # 확대
        scale = min(scale + scale_step, max_scale) #scale이 max_scale 넘지 않게
    elif key == ord('-'):  # 축소
        scale = max(scale - scale_step, min_scale) #scale이 min_scale 넘지 않게

cv2.destroyAllWindows()
```
